# Stabilized ICA for transcriptomic data   

In this jupyter notebook, we use stabilized ICA algorithm to decompose GE matrices of pediatric datsets. We study the extraction of ICA components for datasets of iPC project. Once all datasets of a tumor type have been decomposed, we perform RBH/MNN analysis to identify robust ongoing processes.  

**Note :** To run this jupyter notebook, input must be txt files with dimensions: n genes * m patients. Only cancer type should be defined. 
Please make sure to check:
-the delimitor used in the input txt files (; , \t).
-the column name for genes in input files (important for RBH analysis)

```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import pathlib
import os
import glob
import ntpath
```

```python
#1. Define the data sets to be analyzed
##Datasets are read from folders of pediatric datasets
tumor_type='ES'
folder_to_study='ES'
subfolder='EwingPDX_preprocessed'

data_folder_input = pathlib.Path( os.path.join('../../data', folder_to_study, subfolder))
data_folder_output = pathlib.Path(os.path.join('../sICA_results', tumor_type))
list_files = os.listdir(data_folder_input)
list_files_only = [f for f in glob.glob(os.path.join('../../data', folder_to_study, subfolder,'*.txt'))] #trick to remove folders...
os.makedirs(data_folder_output,exist_ok=True)
```

```python
#2. Perform sICA, plot MSTD and stability and save metagenes/samples in batch

for file in list_files_only:
    file_name_only=ntpath.basename(file)
    print("")
    print("Start analysis for", file_name_only)
    os.makedirs(data_folder_output/file_name_only,exist_ok=True)
    df = pd.read_csv( file, sep="\t", index_col=0)
    print(df.shape)

    X = df.apply(func = lambda x : x - x.mean() , axis = 1)
    X.info()

    #i. Most Stable Transcriptome Dimension (MSTD)
    ##In order to select the number of ICA components, we refer to "Determining the optimal number of independent 
    ##components for reproducible transcriptomic data analysis" Kairov et al. 2017: trade-off between a sufficiently 
    ##large number of components to capture all essential information and a restricted dimension to avoid ending up
    ##with a lot of very unstable components which only capture noise.
    ##To do so, we plot the stability distribution for a wide range of different number of components M = 5, ... ,100 
    ##and we observe which choice of M most satisfies the trade-off.
    from sica.base import MSTD
    MSTD(X.values , m = 2 , M = min(100,len(X.columns)-1) , step = 10 , n_runs = 20 , max_iter = 2000)
    fig_MSTD = plt.gcf() #get the last plot
    fig_MSTD.savefig(data_folder_output/file_name_only/"MSTD_choice.pdf", dpi=300)
    print("")
    print("MSTD done for", file_name_only)

    #ii. Select the nb of components
    ##Define a rule for defining the number of components. Current strategy:
    #Max nb of components: 50
    #Min nb of components: nb of patients/3 -> take floor with int
    dim_opt = min(50, int(len(X.columns)/3))

    #iii. Stabilized ICA
    from sica.base import StabilizedICA 
    start = time.time()
    sICA = StabilizedICA(n_components = dim_opt , max_iter = 2000 , n_jobs = -1)
    sICA.fit(X , n_runs = 100 , plot = True ,normalize = True)
    end = time.time()

    minutes, seconds = divmod(end - start, 60)
    print("running time (min): " + "{:0>2}:{:05.2f}".format(int(minutes),seconds))
    fig_component_stability = plt.gcf()
    fig_component_stability.savefig(data_folder_output/file_name_only/"Stability_ICA_components.pdf", dpi=300)
    print("")
    print("sICA done for", file_name_only)

    #iv. Multidimensional Scaling (MDS)
    ##Some explanation for Multidimensional Scaling can be found on the sklearn page : 
    ##https://scikit-learn.org/stable/modules/generated/sklearn.manifold.MDS.html
    fig_MDS , ax = plt.subplots(figsize = (10 , 7))
    start = time.time()
    sICA.projection(ax = ax , method = 'mds')
    ax.set_title("Multidimensional scaling for ICA components")
    end = time.time() 
    minutes, seconds = divmod(end - start, 60)
    print("running time (min): " + "{:0>2}:{:05.2f}".format(int(minutes),seconds))
    fig_MDS.savefig(data_folder_output/file_name_only/"MDS.pdf", dpi=300)
    print("")
    print("MDS done for", file_name_only)

    #v. Study of metagenes and metasamples
    #Here we save S and A, useful for further analyses and visual inspection.
    Metagenes = pd.DataFrame(sICA.S_ , columns = df.index , index = ['metagene ' + str(i) for i in range(sICA.S_.shape[0])])
    Metasamples = sICA.A_.copy()
    Metasamples.index = df.columns
    Metasamples.columns = ['metasample ' + str(i) for i in range(sICA.A_.shape[1])]    print("Metagenes dim", sICA.S_.shape)
    print("Metasamples dim", sICA.A_.shape)
    transpose_metagenes=Metagenes.T
    transpose_metagenes.to_csv(data_folder_output/file_name_only/"Metagenes.csv", index = True)
    Metasamples.to_csv(data_folder_output/file_name_only/"Metasamples.csv", index = True)

    #We can check that we do obtain sparse ICA components by plotting the distribution of "metagene 1" values 
    ##(i.e the coefficients associated with "metagene 1" for each feature)
    def distribution(Components ,  name , bins):  
        s = Components.loc[name]
        s.hist(bins = bins , weights=np.zeros_like(s) + 1. / s.size , grid = False)
        return

    plt.figure(figsize = (10 , 7))
    distribution(Metagenes , name = 'metagene 1' , bins = 30)
    plt.title("Distribution for metagene 1")
    plt.xlabel("Coefficient value")
    plt.ylabel("Frequency")
    print("")
    print("Done for", file_name_only)
    print("")    
    
    plt.close("all")
```

3. Study of reproducibility - MNN graph

In order to study the reproducibility of ICA components between the four data sets we use the notion of Reciprocal Best Hit (RBH). We consider that a component 𝑖 extracted from a data set D1 and another component 𝑗 extracted from a data set D2 are linked if:

|𝜌𝑖𝑗|=max{|𝜌𝑖𝑘|,∀𝑘∈𝐷2}=max{|𝜌𝑙𝑗|,∀𝑙∈𝐷1}

Here 𝜌𝑖𝑗 represents the Pearson's correlation coefficient between the components 𝑖 and 𝑗. We chose to use Pearson's coefficient over non-parametric coefficients such as Kendall's or Spearman's coefficients since it is strongly influenced by extreme values. Yet in our case these extreme values characterize our ICA components.
**Note :** The layout of the following RBH network is strongly influenced by a random initialisation (Fruchterman-Reingold force-directed algorithm). One will have to run the following result several times until a satisfying result is obtained.

```python
#3. Perform RBH/MNN analysis to identify robust components across datasets of the same cancer type
#Prepare input for MNN analysis
## i. Read metagenes from the saved files

Sources = []
#Read the metagene files
dir=os.listdir(os.path.join('../sICA_results', tumor_type))
#order the directories by alphabetic order, to read the files by alphabetic order -> to sort Sources
dir=sorted(dir)

for dataset in dir:
    path_to_output_folder = pathlib.Path(os.path.join('../sICA_results',tumor_type,dataset,"Metagenes.csv"))
    metagenes = pd.read_csv(path_to_output_folder, sep=",", index_col=0)
    duplicates = metagenes.index.duplicated()
    
    if np.sum(duplicates) > 0 :
        print("yes duplicates!")
        metagenes = metagenes.loc[~duplicates , : ]
        #print( "duplicates : " , list(metagenes.index[duplicates]))
    else :
        print("No duplicates for this data set !")
              
    Sources.append(metagenes.T)
    
#to check if there are duplicated gene names    
#for s in Sources:
    #print("duplicates?",s.columns.duplicated().any())

##ii. Prepare the annotations for the datasets
#Read file names
dataset_infos = pd.read_csv("../../data/datasets_info", sep="\t")
print(dataset_infos.head())
print(type(dataset_infos))
tumor_type_dataset = dataset_infos.loc[dataset_infos["Tumor_type"] == tumor_type]
#order the datasets by alphabetic order, to insure the acronyms will be given to the right datasets
tumor_type_dataset.sort_values(by=['File_name'], inplace=True)
print(tumor_type_dataset)
dataset_names=tumor_type_dataset["Acronym"]
```

```python
## Run MNN with right inputs
color=['k'] * len(dataset_names)

from sica.mutualknn import MNNgraph

fig_MNN , ax = plt.subplots(figsize = (10 , 7))
cg = MNNgraph(data = Sources , names=list(dataset_names) , k=1 , weighted = True)
cg.draw(ax=ax, colors = color , spacing = 2) # ['r' , 'k' , 'b' , 'g', 'yellow', 'purple' , 'brown', 'c' ]
ax.legend(bbox_to_anchor=(1.25 , 1))
fig_MNN.savefig(data_folder_output/file_name_only/"MNNgraph.pdf", dpi=300)

cg.export_json(data_folder_output/"RBH.json")
```
